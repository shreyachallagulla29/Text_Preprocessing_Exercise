```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import spacy

nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

# Load spaCy model
nlp = spacy.load('en_core_web_sm')
```

```python
text = """

Natural Language Processing, also known as NLP, is one of the most important areas of Artificial Intelligence.

Today, companies use NLP in chatbots, search engines, recommendation systems, social media analytics,

customer feedback analysis, automatic translation, and voice assistants.

 

In 2026, many organizations are collecting huge amounts of text data from websites, emails, reviews,

online forms, tweets, product comments, and support tickets. However, this text data is often messy.

For example, a customer may write: "I LOVE this product!!! Visit https://example.com/product123 #Amazing #AI".

Another user may write: "The service was not good... I waited 45 minutes and no one replied."

 

Before applying machine learning or deep learning models, we need to clean and prepare the text.

This step is called text preprocessing. It may include converting text to lowercase, removing URLs,

removing punctuation, removing numbers, tokenizing sentences into words, removing stop words,

stemming, lemmatization, and sometimes removing emojis or special characters.

 

Good preprocessing can improve the quality of NLP models. However, preprocessing should be done carefully.

For example, removing the word "not" from the sentence "I am not happy" can change the meaning completely.

Similarly, removing hashtags may not always be a good idea in social media analytics because hashtags

can contain useful information about topics, opinions, or trends.

 

Your task is to build a complete NLP preprocessing pipeline for this text.

You should clean the text step by step and observe how the output changes after each step.

"""


print("ORIGINAL TEXT:")
print(text)
```

```python
import re

def clean_text_step1(text):

    print("Original text sample:")
    print(text[:200] + "...\n")
    
    # Step 1.1: Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # Step 1.2: Remove only hashtag symbols (#), keep the word
    text = re.sub(r'#', '', text)

    
    # Step 1.3: Remove punctuations (keep letters, numbers, and spaces)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    
    # Step 1.4: Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    
    return text

# Apply cleaning with verbose output
text_step1 = clean_text_step1(text)

print("FINAL CLEANED TEXT")
print(text_step1)
```

```python
# STEP 2: Convert to lowercase
text_step2 = text_step1.lower()
print("STEP 2: After converting to lowercase")
print(text_step2)
```

```python
# STEP 4: Tokenization (NLTK)
tokens_nltk = word_tokenize(text_step2)
print("STEP 4: After tokenization (NLTK)")
print(f"Total tokens: {len(tokens_nltk)}")
print(tokens_nltk)  # Print the 6th token as a sample
```

```python
# STEP 5: Remove stop words (NLTK)
# STEP 6: Remove stop words but KEEP negations
stop_words = set(stopwords.words('english'))

# Define negation words to keep
negations = {'not', 'no'}

# Remove negations from stop words list
stop_words = stop_words - negations

# Filter tokens
tokens_no_stopwords = [word for word in tokens_nltk if word not in stop_words]

print("STEP 6: After removing stop words (NLTK) - Negations KEPT")
print(f"Tokens before: {len(tokens_nltk)}")
print(f"Tokens after: {len(tokens_no_stopwords)}")
print(f"Negations kept: {negations}")
print("\nTokens (showing 50 onwards):")
print(tokens_no_stopwords[50:])

# Show if any negations were found in the text
negations_found = [word for word in tokens_no_stopwords if word in negations]
print(f"\nNegations found in text: {set(negations_found)}")
```

```python
# STEP 7: Lemmatization using NLTK
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
tokens_lemmatized_nltk = [lemmatizer.lemmatize(word) for word in tokens_no_stopwords]

print("STEP 7: After lemmatization (NLTK)")
print(f"\nFirst 50 lemmatized tokens:")
print(tokens_lemmatized_nltk[:50])

```

```python
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Apply lemmatization directly on each token
tokens_lemmatized_spacy = [nlp(token)[0].lemma_ for token in tokens_no_stopwords]


print("spaCy LEMMATIZATION")
print(f"Original tokens (no stopwords): {len(tokens_no_stopwords)}")
print(f"spaCy lemmatized tokens: {len(tokens_lemmatized_spacy)}")

print(f"\nFirst 30 original tokens:")
print(tokens_no_stopwords[:30])

print(f"\nFirst 30 spaCy lemmas:")
print(tokens_lemmatized_spacy[:30])
```

**Comparision between NLTK and Spacy Lemmatization**
* Spacy Lemmatizer worked well compared to the NLTK and the stem is recognized better.


```python
# COMPARISON: NLTK vs spaCy Lemmatization
print("COMPARISON: NLTK vs spaCy")

print(f"\nNLTK lemmatized tokens: {len(tokens_lemmatized_nltk)}")
print(f"Sample: {tokens_lemmatized_nltk[:30]}")

print(f"\nspaCy lemmatized tokens: {len(tokens_lemmatized_spacy)}")
print(f"Sample: {tokens_lemmatized_spacy[:30]}")

# Find unique differences in order of first appearance
print(f"\n{'Original Word':<20} {'NLTK Lemma':<20} {'spaCy Lemma':<20}")
print("-" * 60)

# Use a set to track already seen words, but maintain order
seen_words = set()
unique_differences = []

# Loop through ALL tokens
for i in range(len(tokens_no_stopwords)):
    original = tokens_no_stopwords[i]
    nltk_lemma = tokens_lemmatized_nltk[i]
    spacy_lemma = tokens_lemmatized_spacy[i]
    
    if nltk_lemma != spacy_lemma:
        # Only add if we haven't seen this word before
        if original not in seen_words:
            seen_words.add(original)
            unique_differences.append((original, nltk_lemma, spacy_lemma))

# Display unique differences in order of first appearance
for original, nltk_lemma, spacy_lemma in unique_differences:
    print(f"{original:<20} {nltk_lemma:<20} {spacy_lemma:<20}")

print(f"\nTotal unique differences: {len(unique_differences)}")
print(f"Total tokens compared: {len(tokens_no_stopwords)}")
```
